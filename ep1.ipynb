### Beautified visual:

```python
# raw data:
days = ['0 days', '1 day', '2 days', '3 days', '4 days', '5 days', '6 days', '7 days',
        '0 days', '1 day', '2 days', '3 days', '4 days', '5 days', '6 days', '7 days']
counts = [18, 43, 53, 90, 43, 61, 31, 32,
          1, 8, 9, 53, 102, 84, 68, 47]
types = ['Pre-SRLA-Season', 'Pre-SRLA-Season', 'Pre-SRLA-Season', 'Pre-SRLA-Season', 'Pre-SRLA-Season', 
         'Pre-SRLA-Season', 'Pre-SRLA-Season', 'Pre-SRLA-Season',
         'Post-SRLA-Season', 'Post-SRLA-Season', 'Post-SRLA-Season', 'Post-SRLA-Season', 'Post-SRLA-Season', 
         'Post-SRLA-Season', 'Post-SRLA-Season', 'Post-SRLA-Season']

# building the dataframe
df = pd.DataFrame(zip(days, counts, types))
df.columns = ['Days', 'Number of Students', 'Season Type']
df.head()
```

```python
# calculating percents for y-axis:
sums = df.groupby('Season Type').sum()
percent = []

for i in range(len(df)):
    percentage = df.iloc[i]['Number of Students'] / sums.loc[df.iloc[i]['Season Type']][0]
    percent.append(percentage)
    
df['Percent of Students'] = percent
df.head()
```

```python
# plotting the visual

# plot frame
sns.set(style='white', rc={'axes.facecolor':'white', 'figure.figsize':(12,8),
                           'figure.facecolor':'white', 'patch.linewidth': 1,
                           'axes.linewidth' : 1, 'axes.edgecolor': 'black'});

# plot
ax = sns.barplot(x='Days', y='Number of Students', data=df, palette = ['dimgrey', 'lawngreen'], hue='Season Type',
            edgecolor='black')

# plot elements
plt.xlabel('')
plt.ylabel('Percent of Students', fontproperties=regular, size=20, color = 'black');
plt.title('Days physically active for at least 60 mins per day during an average 7-day week', 
          fontproperties=regular, size=22, color = 'black');
plt.xticks(fontproperties=regular, size=16, color = 'black');
plt.yticks(fontproperties=regular, size=18, color = 'black');
L = ax.legend();
plt.setp(L.texts, family='Consolas');
sns.despine();

plt.savefig('pre_vs_post.png')
```

### General Fact Rings

```python
# rings based off of basic facts:

# completing the marathon:
finish = 99.5
dont = 0.05
total = 100
    
sizes = [finish, dont]
colors = ['lightgreen', 'lightcoral']
    
sns.set(style='white', rc={'figure.figsize':(5,5)})
 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, labels=None, autopct=None, startangle=90,
            wedgeprops={"edgecolor":"black",'linewidth': 3, 'linestyle': 'solid', 'antialiased': True})
    
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white', ec='black', lw=3)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal') 
    
plt.plot([-0.05, -0.1], [-0.6, -0.75], 'k-', lw=3, color = 'black')
plt.text(-0.18, -0.55, str(finish) + '%', fontproperties=regular, size=30, color = 'green');

plt.title('Student Finisher : Non-Finisher Ratio', fontproperties=regular, size=24, color = 'black');
    
plt.tight_layout()
plt.savefig('finisher_ring.png')
```
